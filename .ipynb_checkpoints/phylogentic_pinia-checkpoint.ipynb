```python
import os
import re
import networkx as nx
import matplotlib.pyplot as plt
from pathlib import Path

print("Imports are OK")
```

```python
SRC_DIR = Path(r"C:\Users\squar\xxxxx\project.desktop\xxxxx\apps\twe")
# look for useXStore
STORE_REGEX = re.compile(r"use([A-Z][A-Za-z0-9_]*)Store")

def collect_store_usage(path: Path):
    usage = {}
    for root, _, files in os.walk(path):
        for file in files:
            if file.endswith(".vue") or file.endswith(".ts"):
                file_path = Path(root) / file
                with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
                    content = f.read()
                    matches = STORE_REGEX.findall(content)
                    if matches:
                        #rel_path = file_path.relative_to(path)
                        #usage[str(rel_path)] = list(set(matches)) # too busy
                        rel_path = file_path.relative_to(path).as_posix()
                        short_label = "/".join(rel_path.split("/")[-2:])
                        usage[short_label] = list(set(matches))
    return usage

def build_graph(store_usage: dict):
    G = nx.DiGraph()
    for file, stores in store_usage.items():
        G.add_node(file, type='file')
        for store in stores:
            store_node = f"{store}Store"
            G.add_node(store_node, type='store')
            G.add_edge(file, store_node)
    return G
def draw_graph(G):
    pos = nx.spring_layout(G, seed=42, k=0.7)  # THe smaller the k the tighter the grouping...   ...1.5 to push apart. 
    node_colors = ['lightblue' if G.nodes[n]['type'] == 'file' else 'orange' for n in G.nodes]

    plt.figure(figsize=(16, 12)) # aspect ratio - funny enough! plotlib defaults to inches. Lol
    nx.draw(
        G,
        pos,
        with_labels=True,
        node_color=node_colors,
        node_size=200,
        font_size=8,
        edge_color='gray',
        arrows=True
    )
    plt.title("Pinia Graph!", fontsize=12)
    plt.tight_layout()
    plt.show()


usage = collect_store_usage(SRC_DIR)
graph = build_graph(usage)
draw_graph(graph)

```
